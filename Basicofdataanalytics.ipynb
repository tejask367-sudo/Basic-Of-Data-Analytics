**Basic Of Data Analytics**

1. Define the role of a Data Analyst in your own words. What value do they bring to an organization?
  - A Data Analyst is someone who turns raw data into clear, useful information that helps people make better decisions. They collect, clean, and analyze data, then interpret what the numbers mean in a real-world context. Instead of just reporting figures, they explain why something is happening and what can be done about it.

2. List three tools commonly used by Data Analysts and explain what each is mainly used for.
     1.Microsoft Excel
       Excel is widely used for data cleaning, basic analysis, and quick calculations. Data analysts use it to sort and filter data, create formulas, build pivot tables, and generate simple charts. It’s especially useful for small to medium-sized datasets and fast exploratory analysis.

     2.SQL (Structured Query Language)
       SQL is used to access and manipulate data stored in databases. Data analysts rely on SQL to query large datasets, filter records, join multiple tables, and aggregate data (such as totals or averages). It is essential for retrieving accurate data efficiently from relational databases.

     3.Tableau (or Power BI)
       Tableau is a data visualization tool used to create interactive dashboards and reports. It helps analysts present insights in a clear and visually engaging way, making complex data easier for stakeholders to understand and explore without technical expertise.

Together, these tools help data analysts collect data, analyze it, and communicate insights effectively.

3. Write the end-to-end analytics workflow in the correct order and briefly explain each step.
  - Define the Problem / Business Question
  Clearly understand what the organization wants to know or solve. This step sets the goal and determines what data is needed.

    Data Collection
    Gather relevant data from various sources such as databases, spreadsheets, APIs, surveys, or logs.

    Data Cleaning and Preparation
    Fix errors, handle missing values, remove duplicates, and transform data into a usable format. This ensures the data is accurate and consistent.

    Data Exploration (Exploratory Data Analysis – EDA)
    Analyze the data to find patterns, trends, and outliers using statistics and visualizations. This helps understand what the data is telling you.

    Data Analysis / Modeling
    Apply analytical techniques (such as aggregations, comparisons, or basic models) to answer the business question and generate insights.

    Data Visualization and Reporting
    Present findings using charts, dashboards, and reports that are easy for stakeholders to understand.

    Insights, Recommendations, and Decision-Making
    Interpret the results, explain their impact, and provide actionable recommendations that support business decisions.

    Monitoring and Iteration
    Track results over time and refine the analysis as new data or business needs arise.

This workflow helps ensure that analytics is structured, accurate, and aligned with real business goals.

4. What is Prompt Engineering, and explain any two types of prompts used in Generative AI?
  - Prompt Engineering is the practice of designing and refining the instructions (prompts) given to a generative AI model so that it produces accurate, relevant, and useful outputs. A well-written prompt clearly tells the AI what to do, how to do it, and sometimes the format or style of the response.
    1) Zero-Shot Prompt
    In a zero-shot prompt, the AI is asked to perform a task without being given any examples. It relies entirely on its existing knowledge to generate an answer.
  
    2)Few-Shot Prompt
    In a few-shot prompt, the user provides a small number of examples to show the AI exactly what kind of response is expected.

8. Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.
  - 1. Descriptive Analytics – What happened?

Explanation:
Descriptive analytics focuses on summarizing historical data to understand past performance.

Key Features:

Reports, dashboards, and scorecards

Aggregation of historical data

Example:
Monthly sales reports, website traffic summaries

2. Diagnostic Analytics – Why did it happen?

Explanation:
Diagnostic analytics examines data to identify the root causes of past outcomes.

Key Features:

Drill-down and data discovery

Correlation and variance analysis

Example:
Identifying reasons for a drop in sales in a specific region

3. Predictive Analytics – What is likely to happen?

Explanation:
Predictive analytics uses statistical models and machine learning to forecast future outcomes.

Key Features:

Regression, classification, time-series models

Probability-based forecasts

Example:
Demand forecasting, credit risk prediction

4. Prescriptive Analytics – What should we do?

Explanation:
Prescriptive analytics recommends actions by evaluating different scenarios and constraints.

Key Features:

Optimization algorithms

Simulation and decision models

Example:
Dynamic pricing, supply chain optimization

5. Cognitive / Augmented Analytics – How can systems assist humans?

Explanation:
This stage uses AI and ML to automate data preparation, insight discovery, and explanation.

Key Features:

Natural Language Query (NLQ)

Automated insight generation

Example:
AI-generated insights in BI tools like Power BI, Tableau

6. Generative AI – How can systems create and reason?

Explanation:
Generative AI goes beyond analysis to create new content, code, scenarios, and strategies using large language and multimodal models.

Key Features:

Natural Language interaction

Automated report writing and storytelling

Scenario generation and decision support

Example:
AI-generated business reports, chat-based analytics, synthetic data generation
